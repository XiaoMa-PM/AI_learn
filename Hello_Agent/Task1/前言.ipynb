# 前言
## 教程定位
- 一本从零开始、理论与实战并重的智能体系统构建指南
- 穿透Agent框架的表象，从第一性原理出发，深入其核心架构，理解其经典范式
- 内容聚焦：AI Native Agent，并将其与流程驱动的低代码工程类（Dify、Coze等）区分，教程专注于前者
- 目标：帮助从一名大语言模型的“使用者”，蜕变成一名智能体系统的“构建者”
## 教程结构
- 第一部分（基础篇）： 铺垫AI与LLM的核心知识，简历宏观认识。
- 第二部分（单体篇）： 这是你动手实践的开始。我们将带你从零开始，构建一个功能完备的单体智能。
- 第三部分（高级篇）： 智能体“学会”思考、拥有记忆和工具，并掌握智能体之间的通信协议和评估方法。
- 第四部分（实战篇）： 实战一系列精心设计的综合案例。
- 第五部分（展望篇）： 旅程的终点是新的起点。你将亲手打造属于你的“毕业作品”。

> 学习每部分内容，回头重新修改前言的总结
