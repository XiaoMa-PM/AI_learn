# 第一章 初识智能体
## 引入
自动化工具--Agent的价值--自驱动--区分“自动化”与“自主化”概念
## 定义智能体：
经典定义（功能角度）：传感器、执行器
- 核心在于自主性，基于感知和内部状态进行独立决策
- 举例：自动驾驶汽车；交易算法

LLM驱动新范式（技术角度）：与传统AI不同
- 核心引擎：基于与训练模型为推理引擎
- 知识来源：不再是预定义的知识库，而是从海量数据中间内化和学习到隐式世界模型
- 工作模式：不再是确定性的，而是概率性的、生成式的。它能理解模糊的自然语言指令，并自主地进行规划、调用工具和动态修正

Agent和Workflow的核心区别
- Workflow：静态的、预定义的流程图
- Agent：自主达成目标。以目标为导向，LLM为大脑

## Design & Work
LLM智能体运行核心机制--智能体循环（Agent Loop）
- 感知（preception）：智能体接受来自环境输入信息，也称为观察（Observation）。
- 思考（Thought）：LLM的驱动核心。基于感知与记忆，进行规划（planning）复杂任务拆分为子任务和工具选择（Tool Selection）决定下一步调用哪些工具
- 行动（Action）：智能体通过执行器执行具体行动，如：通过调用一个工具
- 观察（Observation）：行动会改变环境状态，环境会返回一个新的感知/观察
T->A->O->T->A->O->T...闭环不断重复循环，直到任务完成

## Reflection
智能体协作场景：
- 开发者工具（人机协同）
- 主动协作者（目标委托）

构建智能体的Challenge：
- 反应式VS规划式：
    - 反应式：智能体基于当前感知立即做出决策；容易陷入局部最优。
    - 规划式：智能体先进行任务规划，再执行；决策时间和计算成本非常高。
    - 混合式（Hybrid）架构：通过“思考-行动-观察”的微循环，在规划和反应之间取得平衡。
- 符号主义局限在于脆弱性
- 亚符号主义局限在于不透明性（黑盒）
- LLM智能体更多被认为是**神经符号主义（Neruo-Aymbolic）**的一种实践，融合了二者优点，同时生成结构化的规划和API调用（符号）
